### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y, color='red', marker='o', label="Data Points")

# Annotate each point
for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})", fontsize=12, ha='right', va='bottom')

# Add title and labels
plt.title("Scatter Plot with Annotations")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
